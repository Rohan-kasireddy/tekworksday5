```python
import pandas as pd
df=pd.read_csv('task.csv')
df
```

```python
#5 Convert age to numeric and replace invalid values (negative or >100) with NaN
import numpy as np
df['age']=pd.to_numeric(df['age'],errors='coerce')
df.loc[(df['age']<0) | (df['age']>100),'age']=np.nan
df
```

```python
#6  Clean salary by removing currency symbols and converting it to numeric.
df["salary"]=(
    df["salary"]
    .astype(str)
    .str.replace("₹","",regex=False)
    .replace("N/A",np.nan)
)
df["salary"]=pd.to_numeric(df["salary"],errors="coerce")
df
```

```python
# 7 convert join_data into datetime format handling multiple data formats
df["join_date"]=pd.to_datetime(df["join_date"],format="%m-%d-%y",errors="coerce",dayfirst=True)
# join date=02/03/2025
df['joining_year']=df['join_date'].dt.year
df
```

```python
#8. extract joining_year from the cleaned joint_date
df = df.copy()
df.loc[:, "joining_year"] = df["join_date"].dt.year
df
```

```python
#9 standardize department values by trimming spaces and converting to
df['department']=(
    df["department"]
    .str.strip()
    .str.upper()
)
df
```

```python
# 10 Normalize city names to a consistent case and spelling
df['city']=df['city'].str.strip().str.title()
df
    
```

```python
# 11 convert performance score into numeric by mapping text values to numbers
performance_map={'poor':0,'excellent':1}
df['performance_score']=df['performance_score'].map(performance_map)
df['performance_score']=pd.to_numeric(df['performance_score'],errors='coerce')
df
```

```python
# 12 clean exprience_years by converting text values to numeric and removing negatives
df['experience_years']=df['experience_years'].replace("five",5)
df['experience_years']=pd.to_numeric(df['experience_years'],errors='coerce')
df.loc[df["experience_years"]<0,"experience_years"]=np.nan
df
```

```python
#13 convert is active values(yes/no/True/False) into boolean format
map={'yes':True,'no':False,'True':True,'False':False}
df['is_active']=df['is_active'].map((map))
df
```

```python
#14 identify and remove duplicate records based on user_id
df=df.drop_duplicates(subset="user_id")
df
```

```python
#15 Create a clean final DataFrame containing only valid, standardized records
final_df = df.dropna(subset=["user_id", "age", "salary", "join_date"])

print("\nFinal Clean DataFrame Shape:", final_df.shape)
final_df.head()
```
